```python
nota = 3.2
asistencia = 0.8
umbral_nota = 4.0
umbral_asistencia = 0.8

if (nota > umbral_nota) and (asistencia >= umbral_asistencia):
    resultado = "Aprobo y tiene exelencia en asistencia"
elif (nota > umbral_nota) and (asistencia < umbral_asistencia):
    resultado = "Aprobo pero no tiene exelencia en asistencia"
else:
    resultado = "Reprobo"
print(f"Ejemplo 3 - Nota: {nota}, tiene una asistencia del {asistencia *100}% => {resultado}")
```
