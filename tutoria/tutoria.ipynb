Craer un diccionario a partir de listas

```python
ciudades = ["Medellin","Bogota","Cali","Barranquilla","Cartagena"]
comidas_tipicas = ["Bandeja Paisa","Ajiaco","Sancocho","Arepa de Huevo","Ceviche"]

dir_comidas_tipicas = {ciudades[i]: comidas_tipicas[i] for i in range(len(ciudades))}
print(dir_comidas_tipicas)
```
